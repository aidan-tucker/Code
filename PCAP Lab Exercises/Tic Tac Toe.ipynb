```python
from random import randrange

def display_board(board):
    # The function accepts one parameter
    # containing the board's current status
    # and prints it out to the console.

    for row in board:
        print(("+" + "-"*7)*3, end="+\n")
        print(("|" + " "*7)*3, end="+\n")
        print(("|" + " "*3) + row[0] + (" "*3 + "|" + " "*3) + row[1] + (" "*3 + "|" + " "*3) + row[2] + (" "*3 + "|"))
        print(("|" + " "*7)*3, end="+\n")
    print(("+" + "-"*7)*3, end="+\n")


def enter_move(board):
    # The function accepts the board's status
    # asks the user about their move
    # checks the input
    # and updates the board according to the user's decision.

    valid = False
    while not valid:
        try:
            move = int(input("Enter your move: "))
            free = make_list_of_free_fields(board)
            for elem in free:
                if move == elem[0]*3 + elem[1] + 1:
                    board[elem[0]][elem[1]] = "O"
                    valid = True
            if not valid: print("Number must be between 1 to 9 and not already taken, try again")
        except ValueError: print("That wasn't a number, try again")
    return board


def make_list_of_free_fields(board):
    # The function browses the board
    # and builds a list of all the free squares
    # the list consists of types
    # while each tuple is a pair of row and column numbers.

    free = []
    for i in range(len(board)):
        for j in range(len(board)):
            if board[i][j] != "O" and board[i][j] != "X": free.append((i,j))
    return free


def victory_for(board, sign):
    # The function analyses the board's status in order to check
    # if the player using 'O' or 'X's has won the game.

    win = True
    for i in range(3):
        for j in range(3):
            if board[i][j] != sign:
                win = False
                break
        if win: return True
    
    win = True
    for i in range(3):
        for j in range(3):
            if board[j][i] != sign:
                win = False
                break
        if win: return True
    
    win = True
    for i in range(3):
        if board[i][i] != sign:
            win = False
            break
    if win: return True
    
    win = True
    for j in range(3):
        if board[j][2 - j] != sign:
            win = False
            break
    if win: return True

    return False


def draw_move(board):
    # The function draws the computer's move and updates the board.

    try:
        free = make_list_of_free_fields(board)
        index = randrange(len(free))
        board[free[index][0]][free[index][1]] = "X"
    except: print("Something went wrong")
    return board


board = [['1','2','3'],['4','X','6'],['7','8','9']]
display_board(board)

for i in range(4):
    enter_move(board)
    print(victory_for(board, "O"), board)
    if victory_for(board, "O"):
        print(board)
        print("You won!")
        break
    draw_move(board)
    if victory_for(board, "X"):
        print("You lost!")
        break
    display_board(board)
print("FINAL BOARD")
display_board(board)
```
