```python
def easter(x):
    a = x % 19
    b = x % 4
    c = x % 7
    x1 = x//100
    x2 = x//400

    e = (8*x1 + 13) // 25
    f = (x1 - x2 - e + 15) % 30
    g = (x1 - x2 + 4) % 7
    h = (19*a + f) % 30
    i = (2*b + 4*c + 6*h + g) % 7
    val = h + i + 22

    if val > 31:
        if val - 31 == 26: return "APR", val - 38
        elif val - 31 == 25: return "APR", f"{val - 38} or {val - 31}"
        return "APR", val - 31
    return "MAR", val


for i in range(1000): print(i + 1582, easter(i + 1582)[0], easter(i + 1582)[1])
```
